```python
# Step 1 : Import Packages and Dataset

import numpy as np #For Linear Algebra
import pandas as pd #For Data Processing, CSV File
import matplotlib.pyplot as plt #For Visualization


df = pd.read_csv (r'C:\Users\oji\Documents\Dataset Portofolio\olympics.csv')
df.head()
```

```python
# Step 2 : Data Cleaning
# Step 2a : Skipping Rows

olympics_df = pd.read_csv(r'C:\Users\oji\Documents\Dataset Portofolio\olympics.csv', skiprows = 1, skipfooter = 1, header = 0)
olympics_df
```

```python
# Step 2 : Data Cleaning
# Step 2b : Renaming Columns
new_names = {
    'Unnamed: 0': 'Country',
    '№ Summer': 'Summer Olympics',
    '01 !': 'Gold',
    '02 !': 'Silver',
    '03 !': 'Bronze',
    '№ Winter': 'Winter Olympics',
    '01 !.1': 'Gold.1',
    '02 !.1': 'Silver.1',
    '03 !.1': 'Bronze.1',
    '# Total Games': 'Total Games', 
    '01 !.2': 'Total Gold',
    '02 !.2': 'Total Silver',
    '03 !.2': 'Total Bronze'
}

olympics_df.rename(columns = new_names, inplace = True)
olympics_df.head()
```

```python
# Step 2 : Data Cleaning
# Step 2c : Remove Country Code from Country Columns

test = olympics_df['Country'].str.split('(', expand = True)
olympics_df['Country'] = test[0]
olympics_df['Country']
olympics_df.head()

```

```python
# Step 2 : Data Cleaning
# Step 2d : Summary of Olympics Data
olympics_df.describe()
```

```python
# Step 3 : Explanatory Data Analysis
# Step 3a : Find out which country has the most collected medals

df_top5 = olympics_df.nlargest(n=5, columns='Combined total')
df_top5 = df_top5.groupby("Country", axis = 0).sum()
df_top5.head()
df_top5['Combined total'].sort_values(ascending=False).plot(kind='bar',figsize=(10,5),rot=90,color=(0.8,0.8,0.6,0.6))
plt.title('Top 5 Countries of Medalist')
plt.ylabel('Number of medals')
plt.xticks(rotation = 90)
```

```python
# Step 3 : Explanatory Data Analysis
# Step 3b : Find which country has the smallest amount of collected medals

df_btm5 = olympics_df.nsmallest(n=5,columns='Combined total')
df_btm5=df_btm5.groupby("Country",axis=0).sum()
df_btm5.head()

df_btm5['Combined total'].sort_values(ascending=False).plot(kind='bar',figsize=(5,5), rot = 90,color=(0.8,0.2,0.6,0.6))
plt.title('Bottom 5 Countries Medalist')
plt.ylabel('Number of medals')
plt.xticks(rotation=90)
```

```python
# Step 3 : Explanatory Data Analysis
# Step 3c : Find out which country has the most gold medal in winter
winter_df = olympics_df.nlargest(n=5,columns='Gold.1')
winter_df=winter_df.groupby("Country",axis=0).sum()
winter_df.head()
winter_df['Gold.1'].sort_values(ascending=False).plot(kind='bar',figsize=(5,5),rot=90,color=(0.6,0.2,0.5,0.6))

plt.title('Top 5 Countries Winter Gold Medalist')
plt.ylabel('Number of medals')
plt.xticks(rotation=90)
```

```python
# Step 3 : Explanatory Data Analysis
# Step 3d : Find out which country has the most medals in Summer

summer_df = olympics_df.nlargest(n=5,columns='Gold')
summer_df=summer_df.groupby("Country",axis=0).sum()
summer_df.head()
summer_df['Gold'].sort_values(ascending=False).plot(kind='bar',figsize=(5,5),rot=90,color=(0.6,0.2,0.6,0.6))

plt.title('Top 5 Countries Summer Gold Medalist')
plt.ylabel('Number of medals')
plt.xticks(rotation=90)
```
